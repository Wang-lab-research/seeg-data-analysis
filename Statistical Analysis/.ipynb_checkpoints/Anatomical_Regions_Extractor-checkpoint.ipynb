```python
import pandas as pd
import os
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
# List of focus areas:
# caudalanteriorcingulate
# rostralanteriorcingulate
# ctx-lh-insula 
# ctx-rh-insula 
# medialorbitofrontal
# postcentral (S1)
# superiorfrontal (PFC)
# insula

# Of interest areas:
areas_of_interest = ['caudalanteriorcingulate','rostralanteriorcingulate','ctx-lh-insula','ctx-rh-insula','medialorbitofrontal','postcentral','superiorfrontal','insula']
```

```python
parent_data_dir = '../../../George Kenefati/Dan Friedman sEEG Data'

sub_id = 'NY846'+'_elec' # Subject ID im the formate: NY### Only change the ID itself, unless folder naming convention changes
```

```python
def data_extractor(parent_data_dir):
    files_extracted = []
    for file in os.listdir(os.path.join(parent_data_dir, sub_id)):
        if file.endswith('AnatomicalRegions.txt'):
            files_extracted.append(file)
            print(f"Marking {file} for extraction")
        
    return(files_extracted)
            
        
```

```python
def data_frame_maker(file_list):
    for i in range(0, len(file_list)):
        file_name = str(file_list[i])
        data_dir = os.path.join(parent_data_dir, sub_id, str(file_list[i]))
        selected_df = pd.read_csv(os.path.join(data_dir))
        selected_df = selected_df[selected_df['ID'] != '%']        
        selected_df = selected_df[(selected_df['Location'].isin(areas_of_interest)) | (selected_df['Location2'].isin(areas_of_interest)) | (selected_df['Location3'].isin(areas_of_interest))]
        selected_df = selected_df.reset_index(drop = True)
        print(f"Completed Dataframe creation for {file_list[i]}")
        save_dir = str(data_dir) + f"/{file_name[:-4]}-Regions_of_Interest.csv"
        selected_df.to_csv(save_dir)
        
```

```python
processing_list = data_extractor(parent_data_dir)
```

```python
processing_list[0]
```

```python
data_frame_maker(processing_list)
```
